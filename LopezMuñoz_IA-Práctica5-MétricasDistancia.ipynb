#### **1) Importar las bibliotecas necesarias y los datos**

```python
import pandas as pd               # Para la manipulación y análisis de datos
import numpy as np                # Para crear vectores y matrices n dimensionales
import matplotlib.pyplot as plt   # Para la generación de gráficas a partir de los datos
import seaborn as sns             # Para la visualización de datos basado en matplotlib
from scipy.spatial.distance import cdist    # Para el cálculo de distancias
from scipy.spatial import distance
%matplotlib inline
```

```python
BCancer = pd.read_csv('WDBCOriginal.csv')
BCancer
```

```python
BCancer.info()
```

```python
print(BCancer.groupby('Diagnosis').size())
```

```python
Lista1=pd.DataFrame(BCancer)  #DEBEMOS CONEVERTIR EN UNA LISTA PARA MODIFICAR LOS DATOS
Lista1
```

```python
Lista1.drop(["IDNumber"], axis=1,inplace=True)
Lista1.drop(["Diagnosis"], axis=1,inplace=True)
Lista1
#DEBEMOS DEJAR UNICAMENTE NUMEROS PARA PODER CONTINUAR
```

```python
print("DATOS DEPURADOS")
Lista1.info()
```

```python
#Estandarización de datos
#Al estar trabajando con diferentes propiedades y con ello diferentes unidades, es fundamental escalar o normalizar los datos
#para que cada una de las variables contribuyan por igual en el análisis, es decir, no tener un resultado sesgado.

from sklearn.preprocessing import StandardScaler, MinMaxScaler  
estandarizar = StandardScaler()                               # Se instancia el objeto StandardScaler o MinMaxScaler 
MEstandarizada = estandarizar.fit_transform(Lista1)         # Se calculan la media y desviación y se escalan los datos
```

```python
pd.DataFrame(MEstandarizada) 
```

```python
#Matrices de distancia
#Matriz de distancias: Euclidiana
DstEuclidiana = cdist(MEstandarizada, MEstandarizada, metric='euclidean')
MEuclidiana = pd.DataFrame(DstEuclidiana)
```

```python
print(MEuclidiana)
```

```python
print(MEuclidiana.round(3))
```

```python
DstEuclidiana = cdist(MEstandarizada[0:10], MEstandarizada[0:10], metric='euclidean')
MEuclidiana = pd.DataFrame(DstEuclidiana)
print(MEuclidiana) 
```

```python
Objeto1 = MEstandarizada[0]
Objeto2 = MEstandarizada[1]
dstEuclidiana = distance.euclidean(Objeto1,Objeto2)
dstEuclidiana 
#DISTANCIA ENTRE OBJETOS
```

```python
#Matriz de distancias: Chebyshev
DstChebyshev = cdist(MEstandarizada, MEstandarizada, metric='chebyshev')
MChebyshev = pd.DataFrame(DstChebyshev)
print(MChebyshev)
```

```python
DstChebyshev = cdist(MEstandarizada[0:10], MEstandarizada[0:10], metric='chebyshev')
MChebyshev = pd.DataFrame(DstChebyshev)
print(MChebyshev)
```

```python
Objeto1 = MEstandarizada[0]
Objeto2 = MEstandarizada[1]
dstChebyshev = distance.chebyshev(Objeto1,Objeto2)
dstChebyshev
#DISTANCIA ENTRE OBJETOS
```

```python
#Matriz de distancias: Manhattan
DstManhattan = cdist(MEstandarizada, MEstandarizada, metric='cityblock')
MManhattan = pd.DataFrame(DstManhattan)
print(MManhattan)
```

```python
DstManhattan = cdist(MEstandarizada[0:10], MEstandarizada[0:10], metric='cityblock')
MManhattan = pd.DataFrame(DstManhattan)
print(MManhattan)
```

```python
Objeto1 = MEstandarizada[0]
Objeto2 = MEstandarizada[1]
dstManhattan = distance.cityblock(Objeto1,Objeto2)
dstManhattan
#DISTANCIA ENTRE OBJETOS
```

```python
# Matriz de distancias: Minkowski
DstMinkowski = cdist(MEstandarizada, MEstandarizada, metric='minkowski', p=1.5)
MMinkowski = pd.DataFrame(DstMinkowski)
print(MMinkowski)
```

```python
DstMinkowski = cdist(MEstandarizada[0:10], MEstandarizada[0:10], metric='minkowski', p=1.5)
MMinkowski = pd.DataFrame(DstMinkowski)
print(MMinkowski)
```

```python
Objeto1 = MEstandarizada[0]
Objeto2 = MEstandarizada[1]
dstMinkowski = distance.minkowski(Objeto1,Objeto2, p=1.5)
dstMinkowski
#DISTANCIA ENTRE OBJETOS
```
